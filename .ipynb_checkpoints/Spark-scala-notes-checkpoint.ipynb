# What is Spark

###### A fast and general engine for processing large data set
###### Its scalable 
###### its fast - 100x faster than map reduce , based on DAG (Directed acyclic graph) , memory based
###### sql friendly - with spark dataserts and dataframes , lower level API RDD ( resilient distributed dataset)
History - Developed in 2009 by matei zaharia (berkeley lab) open sourced in 2010 2013 Apache software foundation took it up, 2014 main big data project since then fast general purpose cluster framework to overcome map reduce limitatioon at that time.

foundation of spark is RDD read only object split across clusters . RDD processing occurs entierly in memory. It can perform operations like map-reduce,join,filter,aggregation

# spark RDD

rdd - resilent distributed dataset
at core spark is rdd, rows are distributed in different computers and processed in parallel. Resilency spark ensure that!
ex node goes down spark spins new node
key/value rdd
reduceBykey(),groupBykey(), sortBykey(),keys(), values()
sql style joins possible on two key/value rdds
transformation only on values- mapValues(), flatmapValues()

spark context - before creating rdd sc(object ensure rdd resilent and distributed)
sc created rdd , spark shell created sc don't need to explicitely create.

creating rdds -
parallelize , sc.textFile(path), hivectx = HiveContext(sc) --> hivectx.sql(write ur sql here), jdbc, cassandra. hbase,
elasticsearch,json,csv,squence files,object files, various conpressed files

transforming rdds-
map(one row --> one row), flatmap(one row --> many row split),filter,distinct,sample,union,intersection,substract,cartitian

actions --
results back to driver script
collect, count, countByvalue,take (like limit for sample), top , reduce etc
lazy evaluation strategy optimization happens when action called 

spark internal -
stage1 (parallelize) reorganized - filetext (rdd), map()
stage 2 action - countByvalue 

every stage is broken into tasks (distributed scross a cluster ) - tasks are scheduled across cluster and executed to get final results



# Spark sql 

dataframe and datasets -
layer on top of spark core ,dataframe is a dataset of row objects
dataframe does runtime schema infere where as dataset does at complie time 
dataset only available to java, scala sporting compile and not to python(dataframe)

SparkSession require for dataset , sparksession can be created within Sparksession .(similar to database session)

SparkSql exposes a jdbc/odbc server(if you build spark with hive support)

UDF's - import org.apache.spark.sql.function.udf define your udf and use (similar to databases)
datasets works best with structured data, use rdd with unstructured make it schema align then use datasets
Use both rdd+dataset for unstructured data

broadcast variable-
makes variables abailable for each executor for fast look ups, refer to variables locally for small dataset that can get fit in memory

accumulator - 
hit counter updated by executor is shared across all executor and signals other executors about the hit found

cache() , persist() --
use these if you refering to same dataset multiple times to avoid recomputation by spark

# Spark on cluster

How distributed spark cluster works spark figures out cluster manager (spark built in , hadoop yarn , masoos ) and integrated automatically distribute the work of mapper and reducer across cluster available to the enviornment spark driver script is running on master node (driver) ---> communicates to cluster manager distribute out the work thasts in driver script to different executor nodes(workers)
cluster manager responsible for failure of indiviual nodes and collects results from individual nodes to driver scripts
Important configurations -- master 1 - yarn - for running yarn/hadoop cluster 2- hostname:port for connecting to spark stand alone cluster 3- mesos://masternode:port A master in sparkConf will overwrite this

hierarchy
script --> command--> configuration files for spark -- num of executor for Yarn default is 2

--executor-memory(manages how much memory available to each executor , should match with physical memory) shouldn't exceed from individual executor node

--total-executor-cores upper limit on multi cores on virtual machine

EMR -- configures hadoop cluster - yarn (cluster manager) spark is faster than map reduce

.repartition() on dataset and .partionBy() on RDD expensive operations like join , reducebyKey(), GroupBy(), lookUp() etc spark won't distribute on its own results also preserve the partitioning info
